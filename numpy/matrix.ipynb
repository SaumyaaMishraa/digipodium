```python
y =np.random.randint(1,1000,(10,10))
print(y)
```

```python
for i in y:
    print(i)
```

```python
for i in y:
    for j in i:
     if j< 500:
        print(j,end=" ")

#y [y<500] for boo;ean indexing and filtering
```

```python
y [ (y> 250)& (y<750)]# bitwise and is used
```

```python
print ('sum of all values :',y.sum())
print('sum of each row:', y.sum(axis=1))
print('sum of each columns :', y.sum(axis=0))
print('---------------------------------------------')
print('mean of all values :', y.mean())
print('mean of each rows :', y.mean(axis=1))
print('mean of each columns :', y.mean(axis=0))
print('---------------------------------------------')
print('standard deviation of all values:', y.std())
print('standard deviation of each rows:', y.std(axis=1))
print('standard deviation of each column:', y.std(axis=0))
print('---------------------------------------------')
print('min and max of all values:', y.min(),y.max())
print('min and max of each rows:', y.min(axis=1),y.max(axis=1))
print('min and max of each columns:', y.min(axis=0),y.max(axis=0))

```

```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(20,25))    
t= np.linspace(0,20,100)
y= np.sin(t)

plt.plot(t,y)
```

```python
x=np.cos(t)

plt.plot(t,x )
```

```python
z=np.tan(t)
plt.plot(t,z)
```

```python

plt.plot(t,y)
plt.plot(t,x)
plt.grid(False)
```

```python
from tkinter import Label


plt.plot(t,y ,Label='sin(t)', color='yellow', linewidth=3,marker='3', markersize=10)
plt.plot(t,x ,Label='cos(t)', color='hotpink', linewidth=2,linestyle='--')
plt.text(9,0.5,'sin(t)',fontsize=10)
plt.text(6,-0.5,'cos(t)',fontsize=10)
plt.legend()
plt.grid(False)
```
